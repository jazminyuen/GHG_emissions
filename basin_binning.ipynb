```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
file_path = "basins.csv"
basins_df = pd.read_csv(file_path, index_col=False)
basins_df.head()
```

```python
basins_df.columns
```

```python
#specify the columns to sum
cols = ['total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016', 'total_emissions_2015',
       'total_emissions_2014', 'total_emissions_2013', 'total_emissions_2012',
       'total_emissions_2011']
# basins_df[cols]
# #define new column that contains sum of specific columns
basins_df["total_emissions"] = basins_df[cols].sum(axis=1)

# #view updated DataFrame
basins_df
```

```python
# add column that contains average of specific columns
basins_df["avg_emissions"] = basins_df[cols].mean(axis=1)
basins_df
```

```python
basins_df = basins_df[['industry_type_subpart', 'total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016', 'total_emissions_2015',
       'total_emissions_2014', 'total_emissions_2013', 'total_emissions_2012',
       'total_emissions_2011', 'total_emissions', 'avg_emissions']]
basins_df
```

```python
basins_df = basins_df[['industry_type_subpart', 'avg_emissions']]
basins_df
```

```python
industry_counts = basins_df.industry_type_subpart.value_counts()
industry_counts
```

```python
#perform data binning on points variable
basins_df['emissions_bin'] = pd.qcut(basins_df['avg_emissions'], q=3)
```
