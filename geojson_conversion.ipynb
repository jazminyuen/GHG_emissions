```python
import pandas as pd
import numpy as np
```

```python
file_path = "basins.csv"
basins_df = pd.read_csv(file_path)
basins_df.head()
```

```python
geojson = []

for i, x in basins_df.iterrows():
#     print(x["longitude"])
    geojson.append({
    "title":x["facility_name"],
    "geometry":{
        "type":"Point",
        "coordinates":[x["longitude"], x["latitude"]]
    },
    "properties":{
        "City": x["city"],
        "State": x["state"],
        "Zip Code": x["zip_code"],
        "Industry Type Subpart": x["industry_type_subpart"],
        "Basin Number": x["basin_num"],
        "Basin Name": x["basin_name"],
        "Total Emissions": {
            "2020 Emissions": x["total_emissions_2020"],
            "2019 Emissions": x["total_emissions_2019"],
            "2018 Emissions": x["total_emissions_2018"],
            "2017 Emissions": x["total_emissions_2017"],
            "2016 Emissions": x["total_emissions_2016"],
            "2015 Emissions": x["total_emissions_2015"],
            "2014 Emissions": x["total_emissions_2014"],
            "2013 Emissions": x["total_emissions_2013"],
            "2012 Emissions": x["total_emissions_2012"],
            "2011 Emissions": x["total_emissions_2011"]
        }  
    }
})
```

```python
import json
x = json.dumps(geojson)
jsonFile = open("basins.json", "w")
jsonFile.write(x)
jsonFile.close()
```
