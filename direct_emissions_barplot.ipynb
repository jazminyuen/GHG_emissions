```python
import pandas as pd
```

```python
file_path = "Direct_Emitters_cleaned.csv"
direct_emitters_df = pd.read_csv(file_path, index_col=False)
direct_emitters_df.head()
```

```python
direct_emitters_df = direct_emitters_df[['industry_type_sector', 'total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016', 'total_emissions_2015',
       'total_emissions_2014', 'total_emissions_2013', 'total_emissions_2012',
       'total_emissions_2011']]
direct_emitters_df
```

```python
avg_direct_emitters_2020 = direct_emitters_df["total_emissions_2020"].mean()
avg_direct_emitters_2019 = direct_emitters_df["total_emissions_2019"].mean()
avg_direct_emitters_2018 = direct_emitters_df["total_emissions_2018"].mean()
avg_direct_emitters_2017 = direct_emitters_df["total_emissions_2017"].mean()
avg_direct_emitters_2016 = direct_emitters_df["total_emissions_2016"].mean()
avg_direct_emitters_2015 = direct_emitters_df["total_emissions_2015"].mean()
avg_direct_emitters_2014 = direct_emitters_df["total_emissions_2014"].mean()
avg_direct_emitters_2013 = direct_emitters_df["total_emissions_2013"].mean()
avg_direct_emitters_2012 = direct_emitters_df["total_emissions_2012"].mean()
avg_direct_emitters_2011 = direct_emitters_df["total_emissions_2011"].mean()
```

```python
data = {'year': ['2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011'],
        'avg_emissions': [avg_direct_emitters_2020, avg_direct_emitters_2019, avg_direct_emitters_2018, avg_direct_emitters_2017,
                         avg_direct_emitters_2016, avg_direct_emitters_2015, avg_direct_emitters_2014, avg_direct_emitters_2013,
                         avg_direct_emitters_2012, avg_direct_emitters_2011]}
yearly_avg_df = pd.DataFrame(data)
yearly_avg_df
```

```python
yearly_avg_df.sort_values(by=['year'], inplace=True)
yearly_avg_df
```

```python
import matplotlib.pyplot as plt
```

```python
yearly_avg_df.plot.bar(x="year", y="avg_emissions")
```
